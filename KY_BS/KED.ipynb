# 업종 선별

경기, 서울 지역 기업들만으로 분석을 진행

2018년, 2019년, 2020년, 총 3개의 기간을 비교하여 업종의 변화율이 가장 큰 5개의 기업을 선별하겠다. (데이터에 년만 있고 달은 없으므로 코로나 직후, 현재를 나누기 힘들 것 같다)

```python
df_SK = df[(df['CTPRVN_NM']=='서울') | (df['CTPRVN_NM']=='경기')]
```

```python
df= df_SK
```

```python
tot_2019 = len(df[df['STDR_YEAR'] == 2019])
tot_2020 = len(df[df['STDR_YEAR'] == 2020])

ratio = []
for i in set(df['INDUTY_LCLAS_CODE']):
    a = len(df[(df['INDUTY_LCLAS_CODE'] == i ) & (df['STDR_YEAR'] == 2019)])/ tot_2019
    print('2019년 업종당 기업수 비율 : ',i, a*100,"%")
    b = len(df[(df['INDUTY_LCLAS_CODE'] == i ) & (df['STDR_YEAR'] == 2020)])/ tot_2020
    print('2020년 업종당 기업수 비율 : ',i, b*100,"%")
    c = b - a
    ratio.append(c * 100)
    print('2020년과 2019년의 기업수 비율의 변화 : ',i, c*100, "%")
```

```python
ROE_2019
```

이 부분에서도 평균값과 중앙값, 총 두 종류로 나뉜다.

```python
ROE_2019['FSTLTN_AM'].describe()
```

```python
ROE_2019['progress'] = 4
```

```python
for i in ROE_2019.index:
    if ROE_2019['AVRG_AM'][i] < 2.350000:
        ROE_2019['progress'][i] = 1
    if (ROE_2019['AVRG_AM'][i] >= 2.350000) & (ROE_2019['progress'][i] < 10.510000):
        ROE_2019['progress'][i] = 2
    if (ROE_2019['AVRG_AM'][i] >= 10.510000) & (ROE_2019['progress'][i] < 26.565000):
        ROE_2019['progress'][i] = 3
    if ROE_2019['AVRG_AM'][i] >= 26.565000:
        ROE_2019['progress'][i] = 4
```
