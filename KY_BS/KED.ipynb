```python
com_pro = df_SK[df_SK['IEM_NM'] == '영업이익율']
```

```python
com_pro_s = come_pro[come_pro['PRCSS_ENTRPRS_SE_CODE'] == 4]
```

```python
com_pro_s['AVRG_AM'].describe()
```

```python
com_pro_s['class'] = 1
```

```python
for i in com_pro_s.index:
    if com_pro_s['AVRG_AM'][i] < -9.800000e-01:
        com_pro_s['class'][i] = 1
    if (com_pro_s['AVRG_AM'][i] <= 3.360000e+00) & (com_pro_s['AVRG_AM'][i] >-9.800000e-01):
        com_pro_s['class'][i] = 2
    if (com_pro_s['AVRG_AM'][i] <= 7.060000e+00) & (com_pro_s['AVRG_AM'][i] > 3.360000e+00):
        com_pro_s['class'][i] = 3
    if com_pro_s['AVRG_AM'][i] > 7.060000e+00:
        com_pro_s['class'][i] = 4
```

```python
com_pro_s_10 = com_pro_s[com_pro_s['STDR_YEAR']== 2010]
```
