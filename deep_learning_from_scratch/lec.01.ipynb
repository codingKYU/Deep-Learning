```python
class Man:
    def __init__(self,name):
        self.name = name
        print("Initialized!")
    
    def hello(self):
        print("Hello "+ self.name + "!")
        
    def goodbye(self):
        print("Good-bye "+ self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

```python
import numpy as np

X = np.array([[51,55],[14,19],[0,4]])
print(X) # numpy를 통해 리스트를 행렬로 표시가능
```

```python
#인덱스를 통한 numpy안의 값 도출
X[0]
X[0][1]

#for문을 통한 numpy값 도출
for row in X:
    print(row)
```

```python
X = X.flatten() #2차원 배열인 X를 1차원 배열로 변환(평탄화)
print(X)
X[np.array([0,2,4])] #X에서 인덱스 0,2,4값을 도출
```

```python
#특정조건에만 만족하는 numpy값 찾기
X > 15 #X에서 15가 넘는것은 TRUE, 못 넘는것은 FALSE
X[X>15] #TRUE인 값들 추출
```

# matplot 이용해서 그래프 그리기

단순 그래프 그리기 (plt.plot)

```python
import matplotlib.pyplot as plt

X = np.arange(0,6,0.1) #0부터 6까지 0.1간격으로 X를 생성
Y = np.sin(X)

plt.plot(X,Y)
plt.show()
```

그래프의 제목과 각 축의 이름(레이블 표시)

```python
x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label = "sin")
plt.plot(x,y2, linestyle = "--",label = "cos") #그래프의 선을 점선으로 표현
plt.xlabel("x") #x축 이름
plt.ylabel("y") #y축 이름
plt.title('sin&cos')
plt.legend() #matplot이 알아서 그래프를 이쁘게 만들어주는것이라고 이해함
plt.show()
```
